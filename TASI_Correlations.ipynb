```python
# ===================================
# By: Abdullah AlQasim
# Twitter: @jalmood966
# Email: jalmood966 [at] gmail
# ===================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import yfinance as yf
```

```python
start = "2022-01-01"
end   = "2023-07-24"

#tickers = ["3010.SR", "3020.SR", "3030.SR", "3040.SR", "3050.SR", "3060.SR", "3080.SR", "3090.SR"]

tickers = ["2010.SR", "2020.SR", "2060.SR", "2090.SR", "2150.SR", "2170.SR", "2180.SR", "2200.SR"]

colnames = []
```

```python
for ticker in tickers:
    data = yf.download(ticker, start, end)   
    if len(colnames) == 0:
        combined = data[['Adj Close']].copy()
    else:
        combined = combined.join(data['Adj Close'])
    
    colnames.append(ticker)
    combined.columns = colnames
```

```python
plt.yscale("log")

for ticker in tickers:
    plt.plot(combined[ticker], label=ticker)
    
plt.legend(loc="upper right")
plt.show()
```

```python
corr_data = combined.pct_change().corr(method="pearson")
#print(corr_data)

sns.heatmap(corr_data, annot=True, cmap="coolwarm")

plt.show()
```
